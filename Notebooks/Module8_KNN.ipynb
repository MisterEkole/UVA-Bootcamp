```python
import numpy as np 
import pandas as pd 

from sklearn import datasets
```

```python
cancer_dt= datasets.load_breast_cancer()
```

```python
cancer_dt
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test= train_test_split(cancer_dt.data, cancer_dt.target, train_size=0.7)
```

```python
#building the knn model

from sklearn.neighbors import KNeighborsClassifier

n_neighbors=5
knn=KNeighborsClassifier(n_neighbors)
```

```python
knn.fit(X_train, y_train)
```

```python
knn.score(X_test,y_test)
```

```python
X_pred=knn.predict(X_test)
```

```python
#visualing knn classifier

import matplotlib.pyplot as plt 
%matplotlib inline

import seaborn as sns
cmap = sns.cubehelix_palette(as_cmap=True)
f, ax = plt.subplots()
points = ax.scatter(
    X_test[:, 0], X_test[:, 1], c=X_pred, s=50, cmap=cmap
 )
f.colorbar(points)
plt.show()
```
